```
def binary_search(container: List, key: Any, *,  left: int = 0, right: int = 0) -> int:
  if right <= left:
    right = len(container)
  middle = (left + right) // 2 #после = свой код
  value = container[middle]
  if value == key:
    return middle # <your_code_here>
    pass
  elif value > key:
    return binary_search(container, key, left = left, right = middle) # <your_code_here>
    pass
  else:
    return binary_search(container, key, left = middle, right = right)
```
